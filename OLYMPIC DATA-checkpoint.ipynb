```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load data set
athletes = pd.read_csv("I:\Tutorial\python\Data Analytics with Python\OLYMPIC DATA\Athlete_events.csv")
region = pd.read_csv("I:\Tutorial\python\Data Analytics with Python\OLYMPIC DATA\\noc_regions.csv")
region
```

# 1. Merge data

```python
# join dataframe
athletes_df = athletes.merge(region, how="left",on = "NOC")
athletes_df
```

```python
# change column name

athletes_df.rename(columns={"region":"Region","notes":"Notes"},inplace = True)
athletes_df
```

# Top 10 countries participating

```python
top_10_countries_participating = athletes_df["Team"].value_counts().sort_index(ascending=False).sort_values(ascending=False).head(10)
top_10_countries_participating
```

# Bar Plot Top 10 countries participating

```python
plt.figure(figsize=(12,6)) 
plt.title("Overall Participation by Country")
sns.barplot(x = top_10_countries_participating.index, y = top_10_countries_participating, palette="Set2")
```

# Age distribution histrogram of participants

```python
plt.figure(figsize=(15,10)) 
plt.title("Age distribution of participants")
plt.xlabel("Age")
plt.ylabel("Number of participants")
age_median = athletes_df["Age"].median()
plt.hist(athletes_df.Age,bins = np.arange(10,80,2),color = 'blue',edgecolor = "white",)
plt.axvline(age_median,ymax=.95,label = "Age Median",color = "r", linewidth = 4)
# plt.hist(athletes_df.Age,bins = 35, range = (10,80,), color = 'blue',edgecolor = "white")
```
